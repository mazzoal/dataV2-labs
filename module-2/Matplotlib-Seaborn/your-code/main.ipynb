#### From the Fitbit data, we want to visually understand:

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

y= fitbit.groupby("Months_encoded").median()['Steps']

x= fitbit['Months_encoded'].unique()


sns.scatterplot(x=x, y=y, data=fitbit)




```
