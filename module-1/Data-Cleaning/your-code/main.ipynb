#### 4. Import the users table.

```python
data = pd.read_sql_query('SELECT * FROM stats.users', engine)

data
```

#### 5. Rename Id column to userId.

```python
data = data.rename(columns={"Id": "userId"})
data
```

#### 6. Import the posts table. 

```python
data1 = pd.read_sql_query('SELECT * FROM stats.posts', engine)

data1.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
data1 = data1.rename(columns={"Id": "postId", "OwnerUserId": "userId"})

data1
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
columns1 = ['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']
data = data[columns1]

data
```

```python
columns2 =['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']
data1 = data1[columns2]
data1
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#right and left merge

data_merge  = data.merge(data1, left_on = 'userId', right_on = 'postId')
data_merge
```

```python
#inner merge
data_merge3 = pd.merge(data, data1)
data_merge3
```

```python
#outer merge
data_merge2 = pd.merge(data, data1, on = 'userId', how = 'outer')

data_merge2
```
