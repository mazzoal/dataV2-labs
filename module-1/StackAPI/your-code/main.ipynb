#### Import the necessary libraries here:

```python
# your code here
import json
import requests
import pandas as pd
import stackapi
from stackapi import StackAPI
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1580515200, todate=1583020799)
questions

answers = SITE.fetch('answers', fromdate=1580515200, todate=1583020799)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
SITE = StackAPI('stackoverflow')
mostvoted = SITE.fetch('questions', fromdate=1583712000, todate=1583798399, min=5, tagged='python', sort='votes')
mostvoted
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
answers = SITE.fetch('answers/{ids}', ids=[6784, 6473], filter='withbody')
answers
```
